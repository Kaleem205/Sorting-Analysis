Bubble sort

```python
import time
import matplotlib.pyplot as plt

# ider bubble sort ki implementation ho rahi ha

def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]


# Arrays jo question ma thi

Arr1 = list(range(1, 6))
Arr2 = list(range(1, 11))
Arr3 = list(range(1, 51))
Arr4 = list(range(1, 101))

arrays = [Arr1, Arr2, Arr3, Arr4]
input_sizes = [len(arr) for arr in arrays]


# measure kar raha ha average execution time ko

def measure_time(sort_function, arr):
    runs = 5
    total_time = 0
    for _ in range(runs):
        copy_arr = arr.copy()
        start = time.perf_counter()
        sort_function(copy_arr)
        end = time.perf_counter()
        total_time += (end - start)
    return total_time / runs



# collect kar raha data ko bubble sort and print sorted arrays

bubble_times = []

for arr in arrays:
    avg_time = measure_time(bubble_sort, arr)
    bubble_times.append(avg_time)

    # Also print the sorted array

    copy_arr = arr.copy()
    bubble_sort(copy_arr)
    print(f"Sorted array of size {len(arr)}: {copy_arr}")

# Display the average times

print("\nAverage Execution Times:")
for size, time_taken in zip(input_sizes, bubble_times):
    print(f"Input size {size}: {time_taken:.8f} seconds")

# Plotting

plt.figure(figsize=(8, 5))
plt.plot(input_sizes, bubble_times, marker='o', color='blue', label="Bubble Sort")
plt.title("Bubble Sort Performance Analysis")
plt.xlabel("Input Size (N)")
plt.ylabel("Average Execution Time (seconds)")
plt.legend()
plt.grid(True)
plt.show()
```


Bubble sort case :

Graph Description:

The graph illustrates the performance of the Bubble Sort algorithm across different input sizes. The x-axis represents the input size (N), while the y-axis shows the average execution time in seconds. As seen from the plot, the execution time increases as the number of elements in the array grows. For smaller arrays (sizes 5 and 10), the execution time is extremely low and the increase is minimal. However, as the input size reaches 50 and 100, the execution time rises more noticeably. This behavior aligns with the expected time complexity of Bubble Sort, which is O(N), indicating that the sorting time grows quadratically with the size of the input. Overall, the graph clearly demonstrates that Bubble Sort is inefficient for larger datasets, as the time taken to sort increases significantly with input size.

Selection sort

```python
import time
import matplotlib.pyplot as plt


# ider selection sort ki implementation ho rahi ha

def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i+1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]


# Arrays jo question ma thi

Arr1 = list(range(1, 6))
Arr2 = list(range(1, 11))
Arr3 = list(range(1, 51))
Arr4 = list(range(1, 101))

arrays = [Arr1, Arr2, Arr3, Arr4]
input_sizes = [len(arr) for arr in arrays]



# measure kar raha ha average execution time ko

def measure_time(sort_function, arr):
    runs = 5
    total_time = 0
    for _ in range(runs):
        copy_arr = arr.copy()
        start = time.perf_counter()
        sort_function(copy_arr)
        end = time.perf_counter()
        total_time += (end - start)
    return total_time / runs

# collect kar raha data ko selection sort and print sorted arrays

selection_times = []

for arr in arrays:
    avg_time = measure_time(selection_sort, arr)
    selection_times.append(avg_time)

    # Also print the sorted array

    copy_arr = arr.copy()
    selection_sort(copy_arr)
    print(f"Sorted array of size {len(arr)}: {copy_arr}")

# Display the average times

print("\nAverage Execution Times for Selection Sort:")
for size, time_taken in zip(input_sizes, selection_times):
    print(f"Input size {size}: {time_taken:.8f} seconds")

# Plotting

plt.figure(figsize=(8, 5))
plt.plot(input_sizes, selection_times, marker='o', color='green', label="Selection Sort")
plt.title("Selection Sort Performance Analysis")
plt.xlabel("Input Size (N)")
plt.ylabel("Average Execution Time (seconds)")
plt.legend()
plt.grid(True)
plt.show()
```

From the graph, it is evident that as the size of the input array increases, the execution time also increases. For small arrays (sizes 5 and 10), the increase in execution time is very minimal. Insertion sort

```python
import time
import matplotlib.pyplot as plt

# ider insertion sort ki implementation ho rahi ha

def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1

        # Move elements greater than key to one position ahead

        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key



# Arrays jo question ma thi

Arr1 = list(range(1, 6))
Arr2 = list(range(1, 11))
Arr3 = list(range(1, 51))
Arr4 = list(range(1, 101))

arrays = [Arr1, Arr2, Arr3, Arr4]
input_sizes = [len(arr) for arr in arrays]



# measure kar raha ha average execution time ko

def measure_time(sort_function, arr):
    runs = 5
    total_time = 0
    for _ in range(runs):
        copy_arr = arr.copy()
        start = time.perf_counter()
        sort_function(copy_arr)
        end = time.perf_counter()
        total_time += (end - start)
    return total_time / runs



# collect kar raha data ko insertion sort and print sorted arrays

insertion_times = []

for arr in arrays:
    avg_time = measure_time(insertion_sort, arr)
    insertion_times.append(avg_time)

    # Also print the sorted array
    copy_arr = arr.copy()
    insertion_sort(copy_arr)
    print(f"Sorted array of size {len(arr)}: {copy_arr}")



# Display the average times

print("\nAverage Execution Times for Insertion Sort:")
for size, time_taken in zip(input_sizes, insertion_times):
    print(f"Input size {size}: {time_taken:.8f} seconds")

# Plotting

plt.figure(figsize=(8, 5))
plt.plot(input_sizes, insertion_times, marker='o', color='red', label="Insertion Sort")
plt.title("Insertion Sort Performance Analysis")
plt.xlabel("Input Size (N)")
plt.ylabel("Average Execution Time (seconds)")
plt.legend()
plt.grid(True)
plt.show()
```

Merge sort

```python
import time
import matplotlib.pyplot as plt



# ider merge sort ki implementation ho rahi ha


def merge_sort(arr):
    if len(arr) > 1:
        # Finding the middle of the array
        mid = len(arr) // 2
        left_half = arr[:mid]
        right_half = arr[mid:]

        # Recursively divide the arrays
        merge_sort(left_half)
        merge_sort(right_half)

        # Merge the sorted halves
        i = j = k = 0

        # Copy data to temporary arrays L[] and R[]
        while i < len(left_half) and j < len(right_half):
            if left_half[i] < right_half[j]:
                arr[k] = left_half[i]
                i += 1
            else:
                arr[k] = right_half[j]
                j += 1
            k += 1

        # Check if any element was left
        while i < len(left_half):
            arr[k] = left_half[i]
            i += 1
            k += 1

        while j < len(right_half):
            arr[k] = right_half[j]
            j += 1
            k += 1




# Arrays jo question ma thi


Arr1 = list(range(1, 6))
Arr2 = list(range(1, 11))
Arr3 = list(range(1, 51))
Arr4 = list(range(1, 101))

arrays = [Arr1, Arr2, Arr3, Arr4]
input_sizes = [len(arr) for arr in arrays]



# measure kar raha ha average execution time ko


def measure_time(sort_function, arr):
    runs = 5
    total_time = 0
    for _ in range(runs):
        copy_arr = arr.copy()
        start = time.perf_counter()
        sort_function(copy_arr)
        end = time.perf_counter()
        total_time += (end - start)
    return total_time / runs



# collect kar raha data ko merge sort and print sorted arrays

merge_times = []

for arr in arrays:
    avg_time = measure_time(merge_sort, arr)
    merge_times.append(avg_time)

    # Also print the sorted array
    copy_arr = arr.copy()
    merge_sort(copy_arr)
    print(f"Sorted array of size {len(arr)}: {copy_arr}")


# Display the average times

print("\nAverage Execution Times for Merge Sort:")
for size, time_taken in zip(input_sizes, merge_times):
    print(f"Input size {size}: {time_taken:.8f} seconds")



# Plotting

plt.figure(figsize=(8, 5))
plt.plot(input_sizes, merge_times, marker='o', color='blue', label="Merge Sort")
plt.title("Merge Sort Performance Analysis")
plt.xlabel("Input Size (N)")
plt.ylabel("Average Execution Time (seconds)")
plt.legend()
plt.grid(True)
plt.show()
```
